```python
#Q1. What is Flask Framework? What are the advantages of Flask Framework?
'''Flask is a popular and lightweight web framework for Python that allows
developers to quickly build web applications. It is designed to be easy to use, 
flexible, and customizable, making it a popular choice for developing small to 
medium-sized web applications.

The main advantages of using Flask include:

1-Lightweight and Modular: Flask is a lightweight framework, with a small core
and modular design, which makes it easy to customize and extend. Developers can 
choose only the components they need, which can help keep the application small and fast.

2-Easy to Learn and Use: Flask is relatively easy to learn and use, especially for
developers who are familiar with Python. The framework has a simple API and doesn't 
require a lot of boilerplate code, which can help speed up development.

3-Flexible: Flask is highly flexible and can be used to build a wide range of web
applications, from simple static sites to complex web applications with multiple 
layers and databases.

4-Built-in Development Server: Flask comes with a built-in development server,
which makes it easy to test and debug applications during development.

5-Large Community: Flask has a large and active community of developers who 
contribute to the framework and provide support to others. This can be helpful
when seeking advice or solutions to problems.

6-Good Documentation: Flask has comprehensive and well-written documentation,
which makes it easy for developers to get started and learn the framework.'''
```

```python
#Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.
from flask import Flask

app = Flask(__name__)


@app.route("/hello_world")
def hello_world():
    return "<h1>Hello, World!</h1>"
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q3. What is App routing in Flask? Why do we use app routes?
'''App routing in Flask refers to the process of mapping URLs to Python functions,
also known as view functions, that handle the request and return a response. 
In other words, app routing in Flask allows developers to specify which function
should be executed when a user visits a particular URL.

We use app routes in Flask for several reasons, including:

1-Handling user requests: App routing allows developers to map incoming requests
to specific Python functions that handle the request and return a response.

2-Separating concerns: App routing allows developers to separate the concerns of
different parts of the application. For example, one set of routes could be responsible
for handling user authentication, while another set could be responsible for handling 
user-generated content.

3-Creating RESTful APIs: App routing is essential when creating RESTful APIs in Flask.
Each route represents a specific endpoint that the API client can access to perform a 
specific action.

4-Implementing dynamic behavior: App routing allows developers to implement dynamic 
behavior in their applications. For example, a route could take an argument that specifies
the ID of a particular resource, and the corresponding view function would return data
related to that resource.

5-Overall, app routing is a fundamental concept in Flask and is crucial to building 
web applications that are flexible, scalable, and easy to maintain.'''
```

```python
#Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
#Company Name: ABC Corporation
#Location: India
#Contact Detail: 999-999-9999

from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def company_data():
    return '''Company Name: ABC Corporation  ,
             Location: India   ,
              Contact Detail: 999-999-9999'''

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
'''The url_for() function is used in Flask to dynamically build URLs for specific routes. 
This function takes the name of the route as its first argument and any keyword arguments 
for the route's parameters.

Here's an example of how to use the url_for() function in a Flask application:'''
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

with app.test_request_context():
    # using url_for to build a URL for the 'hello' route with a 'name' parameter
    print(url_for('hello', name='Alice')) # outputs: /hello/Alice

    
'''In the example above, we define a route for the /hello/<name> endpoint, where name is
a dynamic parameter. We then use the url_for() function to dynamically build a URL for 
this route with a name parameter set to 'Alice'. The resulting URL is printed to the 
console.

Note that we use the with app.test_request_context() block to create a test request context
for the application, which is required when using the url_for() function outside of a view 
function. This allows Flask to access information about the application context and build 
the correct URL.'''
```
